```python
columns = ['User', 'Card', 'Year', 'Month', 'Day', 'Time', 'Amount', 'Use Chip',
           'Merchant Name', 'Merchant City', 'Merchant State', 'Zip', 'MCC',
           'Errors?', 'Is Fraud?']

bq_columns = [col.lower().replace(' ', '_').replace('?', '') for col in columns]
```

```python
async def recode_columns(file):
    try:
        df = pd.read_csv(file,dtype=str)
        df.columns = bq_columns
        df.to_csv(os.path.join('data','clean',file.split('/')[-1]),index=False)
        return True
    except Exception as e:
        print(f'Error processing {file}: {e}')
        return False
```

```python
await asyncio.gather(*map(recode_columns,files))
```
